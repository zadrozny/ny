# Visualizing New York

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

## the data
N = 5
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Philadelphia']
populations = [8550405, 3971883, 2720546, 2296224, 1567442] 

## necessary variables
ind = np.arange(N)                # the x locations for the groups
width = 0.35                      # the width of the bars

## the bars
rects1 = ax.bar(ind, populations, width,
                color='blue',
                error_kw=dict(elinewidth=2,ecolor='red'))

# axes and labels
ax.set_xlim(-width,len(ind)+width)
ax.set_ylim(0,9000000)
ax.set_ylabel('Population')
ax.set_title('Five most populuous cities in the US')
xTickMarks = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Philadelphia']
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)


plt.show()
```

# Population of New York City within New York

```python
pop_nyc_2015 = 8550405 # factfinder.census.gov/bkmk/table/1.0/en/PEP/2015/PEPANNRES/0400000US36.05000
pop_nys_2015 = 19795791 # (2015 est) http://www.census.gov/popest/data/state/totals/2015/tables/NST-EST2015-01.csv
pop_nyc_2015_perc = round(pop_nyc_2015/pop_nys_2015, 2)*100
pop_nyc_2015_rest = (1 - round(pop_nyc_2015/pop_nys_2015, 2))*100

# The slices will be ordered and plotted counter-clockwise.
labels = 'New York City', 'Rest of New York State'

# sizes = [(pop_nyc_2015/pop_nys_2015)*100]
sizes = [pop_nyc_2015_perc, pop_nyc_2015_rest]
colors = ['yellowgreen', 'lightskyblue']#, 'gold', 'lightcoral']
explode = (0.1, 0)#, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, 
        colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
```
